# Preprocessing

```python
def preprocess(text):
  # convert text into lowercase
  text=text.lower()

  # remove punctuation,numbers from the text
  text=re.sub('[^a-zA-Z]',' ',text)

  #Tokenize: Convert text into list of words
  tokens=word_tokenize(text)

  #Remove stopwords and apply lemmatization
  tokens=[token for token in tokens if token not in stop_words]

  #Lemmatize each token: convert words into its root from
  lemmatize_tokens=[lemmatizer.lemmatize(token) for token in tokens]

  return " ".join(lemmatize_tokens) #converting back to the text format
```

```python
#Checking if the 'preprocess' function works
sample_review="The product quality is really good and worth the price. It arrived on time and works exactly as described. Very satisfied with the purchase."
preprocess(sample_review)
```

```python
cleaned_reviews=[]
for review in reviews:
  cleaned_reviews.append(preprocess(review))
```

```python
cleaned_reviews[:2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(cleaned_reviews,targets,test_size=0.2,
                                               random_state=42,
                                               stratify=targets)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=10000) #setting limit to max number of words
```

```python
x_train_features = vectorizer.fit_transform(x_train)
```

```python
x_train_features.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
multinominal_nb = MultinomialNB()
multinominal_nb.fit(x_train_features,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
y_pred_train = multinominal_nb.predict(x_train_features)
```

```python
accuracy_train=accuracy_score(y_train,y_pred_train)
precision_train=precision_score(y_train,y_pred_train)
recall_train=recall_score(y_train,y_pred_train)
#
```
